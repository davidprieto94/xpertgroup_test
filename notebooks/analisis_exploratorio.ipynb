```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ruta de los datos limpios
cleaned_data_file = "../outputs/cleaned_data.csv"
data = pd.read_csv(cleaned_data_file)

# Resumen estad√≠stico
print("Resumen estad√≠stico:")
print(data.describe())

# Gr√°ficos b√°sicos
# Distribuci√≥n de una variable num√©rica
sns.histplot(data['Price'], kde=True)
plt.title("Distribuci√≥n de Precios")
plt.savefig("../outputs/distribucion_precios.png")
plt.show()

# Correlaciones
correlation = data.corr()
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Matriz de Correlaci√≥n")
plt.savefig("../outputs/matriz_correlacion.png")
plt.show()
```
